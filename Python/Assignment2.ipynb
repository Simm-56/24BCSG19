 Q1. Write a program to input electricity unit charges and calculate total electricity bill according to the given condition:

    For first 50 units Rs. 0.50/unit
    For next 100 units Rs. 0.75/unit
    For next 100 units Rs. 1.20/unit
    For unit above 250 Rs. 1.50/unit
    An additional surcharge of 20% is added to the bill.

```python
unit = int(input("Enter the no of units:"))
if unit <= 50:
    charges = 0.5*50
elif unit <= 150 and unit >= 50:
    charges = (0.5*50)+(0.75*(unit-50))
elif unit <= 250 and unit >= 150:
    charges = (0.5*50)+(0.75*100)+(1.2*(unit-150))
elif unit > 250:
    charges = (0.5*50)+(0.75*100)+(1.2*100)+((unit-250)*100)
total_bill = charges + (charges*0.2)
print("The total electricity cost is:",total_bill)
```

Q6. Write a program for checking the speed of drivers..

	If speed is less than 70, it should print “Ok”.
	Otherwise, for every 5km above the speed limit (70), it should give the driver one demerit point and print the total number of demerit points.For example, if the speed is 80, it should print: “Points: 2”.If the driver gets more than 12 points, it should print: “License suspended”


```python
speed = int(input("Enter the speed of the driver:"))
if speed <= 70:
    print("Okay")
elif speed >= 0:
    extra_speed = speed-70
    demerit = extra_speed//5
    print("Demerit point=",demerit)
    if(demerit>12):
        print("License suspended")
else:
    print("Speed can not be negative.")
```

Q7. A year is a leap year if it is divisible by 4, except that years divisible by 100 are not leap years unless they are also divisible by 400. Write a program that asks the user for a year and prints out whether it is a leap year or not.

Q9. Write a  program to check whether the triangle is equilateral, isosceles or scalene triangle.

```python
side1 = int(input("Enter the first side:"))
side2 = int(input("Enter the second side:"))
side3 = int(input("Enter the third side:"))
if (side1+side2 > side3) and (side2+side3 > side1) and (side1+side3 > side2):
    if (side1==side2==side3):
        print("It is a equilateral triangle.")
    elif (side1==side2) or (side2==side3) or (side3==side1):
        print("It is a isosceles triangle")
    elif (side1!=side2!=side3):
        print("It is a scalene triangle")
else:
    print("Not a valid triangle")
```
