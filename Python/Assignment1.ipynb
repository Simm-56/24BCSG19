**Q16. Find the solution of a quadratic equation.**

```python
a = int(input("Enter the coefficient of x^2: "))
b = int(input("Enter the coefficient of x: "))
c = int(input("Enter the constant coefficient: "))
determinant = (b ** 2 - 4 * a * c) * 0.5
print("First Root:", (-b + determinant) / (2 * a))
print("Second Root:", (-b - determinant) / (2 * a))
```

**Q18. Convert the temperature from celsius to fahrenheit and vice versa.**

```python
fahrenheit = float(input("Enter the temperature in Fahrenheit: "))
print("Equivalent Celsius temperature:", 5 / 9 * (fahrenheit - 32))
celsius = float(input("Enter the temperature in Celsius: "))
print("Equivalent Fahrenheit temperature:", 9 / 5 * celsius + 32)
```

**Q20. Take input of two numbers and perform all logical and bitwise operator.**

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print("Logical NOT:", not num1)
print("Logical AND:", num1 and num2)
print("Logical OR:", num1 or num2)
print("Logical XOR:", (num1 and not num2) or (not num1 and num2))
print("Logical XNOR:", (num1 and num2) or (not num1 and not num2))
print()
print("Bitwise NOT: ", ~num1)
print("Bitwise AND:", num1 & num2)
print("Bitwise OR:", num1 | num2)
print("Bitwise XOR:", num1 ^ num2)
print("Bitwise XNOR:", ~(num1 ^ num2))
print("Bitwise Left Shift:", num1 << num2)
print("Bitwise Right Shift:", num1 >> num2)
```
