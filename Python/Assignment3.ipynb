**Q1. Write a program to find the smallest and largest digit in a number.**

```python
num = int(input("Enter a number:"))
large = 0
small = 9 
while (num>0):
    digit = num%10
    if digit>large:
        large=digit
    elif digit<small:
        small=digit
    num = num//10
print("Smallest:",small)
print("Largest:",large)
```

**Q3.	Write a program to check whether a number is binary or not.**

```python
num = int(input("Enter a number:"))
num_binary=False
temp = num
while(temp>0):
    digit=temp%10
    if digit==0 or digit==1:
        num_binary=True
    else:
        num_binary=False
    temp=temp//10
if (num_binary):
    print("It is a binary number.")
else :
    print("It is not a binary number.")
        
```

**Q4.	Write a program to count how many digits are even and odd in a number.**

```python
num = int(input("Enter a number:"))
even=0
odd=0

while(num>0):
    digit=num%10
    if(digit%2==0):
        even += 1
    else:
        odd +=1
    num//=10
print("There are:",even,"even digits")
print("There are:",odd,"odd digits")       
```

**Q8.	Print numbers from 1 to N, but skip numbers whose digit sum is divisible by 3.**

```python
limit = int(input("Enter the limit"))
for i in range (1,limit+1):
    temp=i
    sum = 0
    while(temp>0):
        sum += temp%10
        temp//=10
    if sum%3 != 0:
        print(i,end=' ')
```

**Q9	Count numbers between 1 and 100 whose reverse is greater than the original number.**

```python
limit = int(input("Enter the limit"))
count = 0
for i in range (1,limit+1):
    temp=i
    rev = 0
   
    while(temp>0):
        rev = (rev*10)+(temp%10)
        temp//=10
    if(rev>i):
        count += 1
print(count)
```

Q13 Swap the first and last digit of a number.

```python
num = int(input("Enter a number: "))

last = num % 10
temp = num
digits = 0


while temp > 0:
    temp //= 10
    digits += 1

first = num // (10 ** (digits - 1))
middle = (num % (10 ** (digits - 1))) // 10

swapped = last * (10 ** (digits - 1)) + middle * 10 + first

print("After swapping:", swapped)
```

16.	Find numbers whose square ends with the same digits as the number itself.(Hit: 5->5^2=25)

```python
limit=int(input("Enter a number"))
for n in range(1,  limit+ 1):
    square = n * n
    temp = n
    count = 0
    while temp > 0:
        temp //= 10
        count += 1
    if square % (10 ** count) == n:
        print(n,end=' ')
```

18.	Print all numbers from 1 to 500 that are divisible by every non-zero digit they contain.

```python
for n in range(1, 501):
    temp = n
    valid = True
    while temp > 0:
        digit = temp % 10
        if digit != 0 and n % digit != 0:
            valid = False
            break
        temp //= 10
    if valid:
        print(n,end=' ')
```
